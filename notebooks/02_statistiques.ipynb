```python
#importation des librairies
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns

#charder le fichier csv
data_std = pd.read_csv("/home/abdoulaye/Documents/Orange Digital center/projets/projet_data/data/ai_student_impact_dataset.csv")
data_std.head(10)
```

```python
# on isole colonne avec les nombres
df_num = data_std.select_dtypes(include='number')

# Calcule les quartilles et percentilles uniquement pour les colonnes contenant des nombres
quartiles_df = df_num.quantile([0.10, 0.25, 0.50, 0.75, 0.90])
display(quartiles_df)
```

```python
# Calcul de l'IQR
iqr = (df_num.quantile(0.75) - df_num.quantile(0.25) )
print(iqr)
```

```python
# === ANALYSE COMPLETE : Weekly_GenAI_Hours ===

# Q1, Q2, Q3
Q1 = data_std["Weekly_GenAI_Hours"].quantile(0.25)
Q2 = data_std["Weekly_GenAI_Hours"].quantile(0.50)
Q3 = data_std["Weekly_GenAI_Hours"].quantile(0.75)

# Statistiques de base
moyenne = data_std["Weekly_GenAI_Hours"].mean()
mediane = data_std["Weekly_GenAI_Hours"].median()
ecart_type = data_std["Weekly_GenAI_Hours"].std()
variance = data_std["Weekly_GenAI_Hours"].var()

# Coefficient de variation
cv = (ecart_type / moyenne) * 100

# IQR
iqr = Q3 - Q1

# Bornes outliers
limite_basse = Q1 - (1.5 * iqr)
limite_haute = Q3 + (1.5 * iqr)

# Détection outliers
outliers = data_std[data_std["Weekly_GenAI_Hours"] > limite_haute]
pourcentage_outliers = (len(outliers) / len(data_std)) * 100

# Affichage
print(" === ANALYSE COMPLETE : Weekly_GenAI_Hours ===")
print(f"Moyenne     : {moyenne:.2f}h")
print(f"Médiane     : {mediane:.2f}h")
print(f"Écart-type  : {ecart_type:.2f}h")
print(f"CV          : {cv:.2f}%")
print(f"Q1={Q1:.2f} | Q2={Q2:.2f} | Q3={Q3:.2f} | IQR={iqr:.2f}")
print(f"Borne haute : {limite_haute:.2f}h")
print(f"Outliers    : {len(outliers)} étudiants ({pourcentage_outliers:.2f}%)")
```
